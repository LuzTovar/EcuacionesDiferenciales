#### En este ejercicio se analiza el comportamiento para el modelo del crecimiento de la población de Malthus

$ P(0)=1000 $

$P(1)=2000$

#### solución
$$P(t) = 1000 e^{Ln 2 (t)}$$

```python
%matplotlib inline

# librerias
import matplotlib.pyplot as plt
import numpy as np
import sympy 

import pandas as pd


from scipy import integrate
from math import log

# imprimir con notación matemática.
sympy.init_printing(use_latex='mathjax')
```

```python
# generar los valores para t
t = np.arange(0, 5, 0.5)
#t
#s=np.exp(log(2)*t)
#s

#e=np.exp(1)
#e

# generar el valor de y de la Ec. diferencial
s = 1000 * (np.exp((log(2) * t)))

#s
```

```python
# ver los datos
d=(t,s)
df = pd.DataFrame(d)
df
```
