### PROBLEMA
Supongamos que hubo un asesinato y la policía llega a la escena del crimen a las 4:00 am. El forense toma la temperatura de la víctima y encuentra que es de 35° C; Luego vuelve a tomar la temperatura una hora después y la lectura marca 33.1° C.                                                     Nota: La temperatura normal promedio de un ser humano = 37                                                                                            Suponiendo que ese día el clima estuvo templado y el ambiente hubo una temperatura constante de 18°C.¿Cuál fue la hora del asesinato?  

### DATOS
T(0)= 35° C,  4:00 am

T(1)= 33.1° C,  5:00 am

T(t)= 37° C

Tm= 18° C

### PROCEDIMIENTO
#### Fórmula General

$T=Ce^{kt}+Tm$

$f(x)= x^2+e^{2t}$

$T(0)=Ce^{k0}+18=35$

$T(0)=C=17$

**$T(1)= 33.1° C,  5:00 am$**

$T(1)=17e^{k0}+18=33.1$

$T(1)=17e^{k1}=15.1$

$T(1)=e^{k1}=\frac{15.1}{17}$

$T(1)=ln⁡ e^{k1} =ln⁡ \frac{15.1}{17} $

$T(1)=k=ln⁡\frac{15.1}{17}$

#### Fórmula Particular


$T(t)=17e^{ln⁡\frac{15.1}{17} t}+18$

$T(t)= 37° C$ 

$17e^{ln⁡\frac{15.1}{17}t}+18=37$

$17e^{ln⁡\frac{15.1}{17}t}=37-18$

$17e^{ln⁡\frac{15.1}{17}t}=19$

$e^{ln⁡\frac{15.1}{17}t}=\frac{19}{17}$

$ln⁡ e^{ln⁡\frac{15.1}{17}t} =ln⁡\frac{19}{17}$

$ln⁡\frac{15.1}{17}t=ln⁡\frac{19}{17}$

$t=\frac{ln⁡\frac{15.1}{17}}{ln⁡\frac{19}{17}}$

### Respuesta: -0.9384= 56 min
